# converting the datatypes

```python
df['dob']=pd.to_datetime(df['dob'],errors='coerce')
df['account_open_date']=pd.to_datetime(df['account_open_date'],errors='coerce')
```

```python
df.dtypes
```

# standardize the gender values

```python
df['gender']=df['gender'].replace({'M':"Male","F":"Female","O":"Others"})
```

```python
df
```

# Clean and Normalize Text Columns

```python
for col in ['city','state','occupation','account_type','loan_status']:
    df[col]=df[col].str.strip().str.title()
```

```python
df
```

# validate Logical data

```python
df=df[df['dob']<df['account_open_date']]
df=df[df['balance']>=0]
df=df[(df['credit_score']>=300)&(df['credit_score']<=850)]
```

# adding Columns 

```python
df['age']=(pd.Timestamp('today')-df['dob']).dt.days//365
df['account_age']=(pd.Timestamp('today')-df['account_open_date']).dt.days//365
df['full_name']=df['first_name']+" "+df['last_name']
df
```

# Reorder and Rename Columns

```python
df = df[['customer_id', 'first_name', 'last_name', 'full_name', 
         'gender', 'dob', 'age', 'city', 'state', 'occupation', 'account_type',
         'account_open_date', 'account_age', 'balance',
         'credit_score', 'loan_status']]
```

```python
df
```

# Ensure column Nmaing Consistency

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

# Exporting the Cleaned dataset

```python
df.to_csv("C:\Banking_project\cleaned_customers.csv",index=False)
```
