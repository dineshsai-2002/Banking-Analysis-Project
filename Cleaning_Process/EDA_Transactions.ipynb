```python
# converting the Datatypes
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors='coerce')
```

```python
df.dtypes
```

```python
# Filling the missing values in the Merchant Column
```

```python
def fill_merchant(row):
    if pd.notnull(row['merchant']):
        return row['merchant']
    elif row['transaction_type'].lower() == 'deposit':
        return 'Bank Deposit'
    elif row['transaction_type'].lower() == 'withdrawal':
        return 'ATM Withdrawal'
    elif row['transaction_type'].lower() == 'upi':
        return 'UPI Transfer'
    else:
        return 'Self Transaction'

df['merchant'] = df.apply(fill_merchant, axis=1)
```

```python
df.isnull().sum()
```

```python
# Validate Amounts
```

```python
df = df[df['amount'] > 0]
```

```python
# Creating New columns for Data Analysis
```

```python
df['year'] = df['transaction_date'].dt.year
df['month'] = df['transaction_date'].dt.month_name()
df['day_of_week'] = df['transaction_date'].dt.day_name()
df['hour'] = df['transaction_date'].dt.hour
```

```python
df
```

```python
# Rename colunms for consistency
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Categorize transaction type (credit/debit)
```

```python
df['transaction_category'] = df['transaction_type'].replace({
    'Deposit': 'Credit',
    'Withdrawal': 'Debit',
    'Bill Payment': 'Debit'
})
```

```python
df
```

```python
# Reorder Columns for Better Structure
```

```python
ordered_cols = [
    'transaction_id', 'customer_id', 'transaction_date', 'year', 'month', 'day_of_week', 'hour',
    'transaction_type', 'transaction_category', 'transaction_mode',
    'merchant', 'amount', 'status_transaction'
]
df = df[ordered_cols]
```
