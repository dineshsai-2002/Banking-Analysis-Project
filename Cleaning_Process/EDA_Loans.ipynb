```python
# Standardize the Text columns
```

```python
df['loan_type'] = df['loan_type'].str.strip().str.title()
df['status'] = df['status'].str.strip().str.title()
```

```python
df
```

```python
# Coverting the Date columns
```

```python
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')
```

```python
df.dtypes
```

```python
# Ensure numeric columns are correct
```

```python
numeric_cols = ['loan_amount', 'interest_rate', 'tenure_months']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

```python
# verify the tenure consistency
```

```python
df['calculated_tenure'] = ((df['end_date'] - df['start_date']).dt.days / 30).round()
df.loc[abs(df['tenure_months'] - df['calculated_tenure']) > 1, 'tenure_months'] = df['calculated_tenure']
df.drop(columns=['calculated_tenure'], inplace=True)
```
